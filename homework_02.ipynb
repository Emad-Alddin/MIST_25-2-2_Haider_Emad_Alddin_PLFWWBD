Если мы хотим, чтобы вывод был на той же строке, мы можем использовать инструкцию `end(" ")` внутри инструкции `print`.

```python
#### shopping_list=[]
while True:
    print("Менеджер списка покупок")
    print("_"*20)
    print("1.Добавить")
    print("2.Показать")
    print("3.Выйти")
    choice= input("Введите номер варианта 1, 2 или 3").strip()
    if choice =="1":
        product=input("Введите название продукта:").strip()
        if product:
            shopping_list.append(product)
            print("Товар добавлен в список.")
        else:
            print("Вы не ввели название продукта.")
    elif choice == "2":
        if len(shopping_list)==0:
            print("Список пуст")
        else:
            for i, item in enumerate (shopping_list, start=1):
                print(f"{i}.{item}")
    elif choice=="3":
         print("Программа завершена.")
         break
    else:
        print("Неизвестная команда, попробуйте снова.")
    
                  
            
            
        
            
            
    
```
