```python
def count_word_frequency (text):
    text=text.lower()
    punctuation = [".", ",", "!", "?", ":", ";", "-", "،", "؛"]
    for sign in punctuation:
        text= text.replace(sign," ")
    words= text.split()
    freq_dict={}
    for word in words:
        freq_dict [word]=freq_dict.get(word,0)+1
    return freq_dict
```

```python
text= "Емад,Акрам,Рим!Рим студентка в Мисис и Емад студент В мисис и Акрам друг Емад "
frequency_dict= count_word_frequency (text)
print (frequency_dict)
```

```python
def calculate_price (price, discount=0, tax_rate=0.2):
    price_after_discount = price * (1 - discount / 100)
    final_price = price_after_discount * (1 + tax_rate)
    return round (final_price,2)
```

```python
price=3000
final= calculate_price (price, discount=0, tax_rate=0.2)
print(final)
```

```python
import datetime
import sys

def analyze_log(file_path):
         Returns empty dict if file not found.
    
    result = {
        'counts': {'INFO': 0, 'WARNING': 0, 'ERROR': 0},
        'time_range': None
    }
    
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
    except FileNotFoundError:
        print(f"File not found: {file_path}", file=sys.stderr)
        return {}
    except Exception as e:
        print(f"Error reading file: {e}", file=sys.stderr)
        return {}
    
    min_date = None
    max_date = None
    
    for line_num, line in enumerate(lines, 1):
        line = line.strip()
        if not line:  # Skip empty lines
            continue
        
        try:
            # Split the line into parts
            parts = line.split(',')
            if len(parts) < 3:
                print(f"Warning: Line {line_num} has invalid format: '{line}'", file=sys.stderr)
                continue
            
            # Parse date and level
            datetime_str = parts[0].strip()
            level_str = parts[1].strip().upper()
            # message = parts[2].strip()  # Not needed for this task
            
            # Parse datetime
            dt = datetime.datetime.strptime(datetime_str, "%Y-%m-%d %H:%M:%S")
            
            # Update level count
            if level_str in result['counts']:
                result['counts'][level_str] += 1
            else:
                # Handle unknown log levels
                result['counts'][level_str] = 1
            
            # Update min and max dates
            if min_date is None or dt < min_date:
                min_date = dt
            if max_date is None or dt > max_date:
                max_date = dt
                
        except ValueError as e:
            print(f"Warning: Line {line_num} has invalid datetime format: '{line}'", file=sys.stderr)
            continue
        except IndexError as e:
            print(f"Warning: Line {line_num} has missing parts: '{line}'", file=sys.stderr)
            continue
        except Exception as e:
            print(f"Warning: Line {line_num} has unexpected error: {e}", file=sys.stderr)
            continue
    
    # Set time range if we have valid dates
    if min_date is not None and max_date is not None:
        result['time_range'] = (
            min_date.strftime("%Y-%m-%d %H:%M:%S"),
            max_date.strftime("%Y-%m-%d %H:%M:%S")
        )
    
    return result


    
   
```

```python
 analysis = analyze_log('log.txt')
 print("Log Analysis Result:")
print(analysis)
    
   
print("\nTesting with non-existent file:")
analysis2 = analyze_log('non_existent_file.txt')
print(analysis2)

       
```
